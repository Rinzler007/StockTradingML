```python
import pandas as pd
import ta
```

```python
tsla = pd.read_csv("TSLA.csv", usecols=['Close'])
tsla['RSI'] = ta.momentum.RSIIndicator(tsla['Close'], window=7).rsi()
tsla['BBBuy'] = ta.volatility.BollingerBands(tsla['Close'], window=80, window_dev=1.9).bollinger_lband()
tsla['BBSell'] = ta.volatility.BollingerBands(tsla['Close'], window=30, window_dev=1).bollinger_hband()
tsla['MACDLine'] = ta.trend.MACD(tsla['Close']).macd()
tsla['MACDSignal'] = ta.trend.MACD(tsla['Close']).macd_signal()
tsla['SMA5'] = ta.trend.SMAIndicator(tsla['Close'], window=5).sma_indicator() 
tsla['SMA15'] = ta.trend.SMAIndicator(tsla['Close'], window=15).sma_indicator() 
tsla['SMA30'] = ta.trend.SMAIndicator(tsla['Close'], window=30).sma_indicator() 
tsla['SMA40'] = ta.trend.SMAIndicator(tsla['Close'], window=40).sma_indicator()
tsla['SMA100'] = ta.trend.SMAIndicator(tsla['Close'], window=100).sma_indicator()
tsla
```

```python
tsla.to_csv("NewFileTSLA")
```
