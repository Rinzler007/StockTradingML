```python
import pandas as pd
import chefboost as chef
```

```python
tsla = pd.read_csv("NewFileTSLA.csv")
tsla
```

```python
df = pd.DataFrame()
```

```python
RSI = []
BB = []
MACD = []
SMA1 = []
SMA2 = []
Closing = []
Decision = []
for i in range(100, 2963):
    Closing.append(tsla.iloc[i,1])
    Decision.append(tsla.iloc[i,12])
    if tsla.iloc[i,2] > 75:
        RSI.append("Sell")
    elif tsla.iloc[i,2] <25:
        RSI.append("Buy")
    else:
        RSI.append("Hold")
        
    if tsla.iloc[i,3] > tsla.iloc[i,1]:
        BB.append("Buy")
    elif tsla.iloc[i,4] > tsla.iloc[i,1]:
        BB.append("Sell")
    else:
        BB.append("Hold")
        
    if tsla.iloc[i,5] > tsla.iloc[i,6]:
        MACD.append("Buy")
    else:
        MACD.append("Sell")
        
    if tsla.iloc[i,9] > tsla.iloc[i,11]:
        SMA1.append("Buy")
    else:
        SMA1.append("Sell")
    
    if tsla.iloc[i,8] > tsla.iloc[i,10]:
        SMA2.append("Buy")
    else:
        SMA2.append("Sell")
        
df["Closing"] = Closing
df["RSI"] = RSI
df["MACD"] = MACD
df["BB"] = BB
df["SMA1"] = SMA1
df["SMA2"] = SMA2
df["Decision"] = Decision
```
